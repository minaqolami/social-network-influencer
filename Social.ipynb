```python
def remove_last_character(value):
    last_char = value[-1]
    if last_char in ["K", "M"]:
        return float(value[:-1])
    else:
        return float(value)
```

```python
dfnew[' Subscribers'] = dfnew[' Subscribers'].apply(lambda  x: remove_last_character(x))
dfnew['Avg Comments']= dfnew['Avg Comments'].apply(lambda  x: remove_last_character(x))
dfnew['Avg. likes'] = dfnew['Avg. likes'].apply(lambda  x: remove_last_character(x))
dfnew['Avg. views\r\n']= dfnew['Avg. views\r\n'].apply(lambda  x: remove_last_character(x))
```
